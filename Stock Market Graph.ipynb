# 2. Visualizing Tesla's The Stock Price Data

Finding the price movement over time (monthly in last 1 year) to get the exact moment when the price went up or down. Graph from matplotlib will help to get when it was, and match it with Twitter dataset before.

```python
import pandas as pd
import numpy
import datetime
import matplotlib.pyplot as plt
import seaborn as sns

TSLA=pd.read_csv('TSLA (2).csv')


plt.figure(figsize=(15,15))
plt.plot("Date", "Close", data=TSLA)
plt.title('Tesla stock price history')
plt.ylabel('Close (USD)')
plt.xlabel('Month')
plt.legend(['Close Price'], loc='upper left')
plt.show()
```
